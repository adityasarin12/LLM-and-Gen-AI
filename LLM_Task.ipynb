```python
import google.generativeai as genai
from google.colab import userdata
genai.configure(api_key=userdata.get('GOOGLE_API_KEY_1'))
```

```python
def analyze_essay(text):
    model = genai.GenerativeModel("gemini-flash-lite-latest")
    base_prompt = f"""
    You are an AI-powered assistant that provides personalized, constructive feedback on student essays.
    You should highlight and return helpful feedback for grammar issues, weak arguments, awkward phrasing,
    clarity, argument structure, and tone and suggest improvement.
    {text}
    """
    response = model.generate_content(base_prompt)

    analysis = response.text.strip()
    return analysis
```

```python
text=''' Technology has improved people’s life in many ways, but sometimes it also make things worse.
     For example, social media helps people connect, but also it cause loneliness when people compare
     themselves with others too much. Also, some argue that technology always leads to progress, but
     this is not true because many people lose jobs to machines and automation. On the other hand,
     education become easier because of online resources, yet some students just copy information
     without thinking critically.In addition, there is a problem with how fast technology change.
     Governments and laws often cannot keep up, so problems like privacy, hacking, and fake news
     spread quickly. While some say we should just adapt, not everyone can adapt at same speed.
     Therefore, it’s necessary to balance benefits and harms, but this balance is hard to define.
     In conclusion, technology is both good and bad, and society should think carefully.
'''
print(analyze_essay(text))
```
